# Data Exploration

The purpose of this notebook is to explore the input datasets and cleanup the data as necessary to produce a better driving model.

```python
import csv
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


def ReadCSVFile(filename):
    lines = []
    with open(filename) as csvfile:
        reader = csv.reader(csvfile)
        for line in reader:
            lines.append(line)
    return lines
```

## Plot the distribution of angular data
This/these plot/plots will show you the number of samples of angular data taken from the driving simulator

```python
lines = ReadCSVFile('data/driving_log.csv')
lines2 = ReadCSVFile('data2/driving_log.csv')

#plot lines over 
measurements = []
speeds = []
for line in lines:
    measurement = float(line[3])
    speed = float(line[6])    
    if measurement != 0 and measurement !=-1 and measurement !=1 and speed >0.5:
        measurements.append(measurement)
        measurements.append(-measurement)
        measurements.append(measurement+0.2)
        measurements.append(-(measurement+0.2))
        measurements.append(measurement-0.2)
        measurements.append(-(measurement-0.2))         
        speeds.append(speed)
        speeds.append(speed)  
        speeds.append(speed)
        speeds.append(speed)
        speeds.append(speed)
        speeds.append(speed)  
        
#for line in lines2:
#    measurement = float(line[3])
#    speed = float(line[6])      
#    if measurement != 0 and measurement !=-1 and measurement !=1 and speed >0.5:  
#        measurements.append(measurement)
#        measurements.append(-measurement)
#        measurements.append(measurement+0.2)
#        measurements.append(-(measurement+0.2))
#        measurements.append(measurement-0.2)
#        measurements.append(-(measurement-0.2))         
#        speeds.append(speed)
#        speeds.append(speed)  
#        speeds.append(speed)
#        speeds.append(speed)
#        speeds.append(speed)
        speeds.append(speed)
        
print(np.size(measurements))
    
hist_train, bins = np.histogram( measurements, bins=21)
print(np.size(hist_train))
bin_max = np.where(hist_train == hist_train.max())
max_bin_val = hist_train.max()
print(max_bin_val)


# Plot the data
fig, ax = plt.subplots()
width = 0.7 * (bins[1] - bins[0])
center = (bins[:-1] + bins[1:]) / 2
set_train = ax.bar( center, hist_train, align='center', width=width)

# add some text for labels, title and axes ticks
ax.set_xlabel('Counts')
ax.set_title('Training Set\nCounts at each angle')


plt.show()


```
